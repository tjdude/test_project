```python
import numpy as np
import os
from matplotlib import pyplot as plt

### Tensorflow 2.0 ###
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras import Input
from tensorflow.keras import layers
from tensorflow.keras import models
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import losses
from tensorflow.keras import optimizers
from tensorflow.keras import metrics
from tensorflow.keras import regularizers
from tensorflow.keras import utils
```

```python
# 아래 있는 model의 plot을 보고 그대로 CNN 모델을 구현하시오
# Hint : 각 layer에서 사용된 이름이 좌측에 적혀있습니다.
#        오른쪽의 2개는 순서대로 input과 ouput의 shape입니다.
# 모델을 구현한 후에 model.summary() 를 써서 맨 아래 있는 것처럼 출력되도록 하시오.

first_layer = layers.Conv2D(shape=(128, 128, 16))
second_layer = layers.MaxPool2D(filters=8, kernel_size=(3, 3), activation=None, padding='same')(first_layer)
third_layer = layers.Conv2D()(second_layer)
fourth_layer = layers.MaxPool2D('relu')(third_layer)
fifth_layer = layers.Conv2D(strides=(2, 2))(fourth_layer)
sixth_layer = layers.MaxPool2D()(fifth_layer)
seventh_layer = layers.Dense(100, activation = 'relu')(sixth_layer)
eighth_layer = layers.Dropout(0.25)(seventh_layer)
final_layer =  layers.Dense(10, activation='sigmoid')(eighth_layer)
model = Model(first_layer, final_layer)
model.summary()
```

```python
tf.keras.utils.plot_model(model, show_shapes=True)
```
